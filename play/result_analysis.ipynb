```python
import pandas as pd
import json
from pathlib import Path
```

```python
contents = Path("log_extracts.txt").read_text()
data = json.loads(contents)

df_logs = pd.DataFrame(data).set_axis(
    ["layer1", "layer2", "layer3", "layer4", "layer5", "layer6", "layer7", "layer8"],
    axis=1,
)


contents = Path("get_maps.json").read_text()
data = json.loads(contents)

df_layers = pd.json_normalize(
    data["result"],
    None,
    meta=["id", "environment", "game_mode", ["map", "id", "name"]],
).drop(
    [
        "pretty_name",
        "image_name",
        "map.name",
        "map.tag",
        "map.pretty_name",
        "map.shortname",
        "map.allies.name",
        "map.allies.team",
        "map.axis.name",
        "map.axis.team",
        "map.orientation",
    ],
    axis=1,
).rename(columns={"map.id": "map_id"})

rename_columns=["game_mode", "attackers", "environment", "map_id"]
for i in range(1,9):
    layer_col = f"layer{i}"
    columns={col: f"layer{i}_{col}" for col in rename_columns}
    df_logs = df_logs.join(df_layers.set_index("id"), on=layer_col).rename(columns=columns)



df_logs["warfare_count"] = df_logs.eq("warfare").sum(axis=1)

df_logs
```
